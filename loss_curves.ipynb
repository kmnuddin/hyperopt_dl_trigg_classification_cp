```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
train_acc_1vs3 = pd.read_csv('train_time_data/1vs3_run-train-tag-epoch_accuracy.csv')
val_acc_1vs3 = pd.read_csv('train_time_data/1vs3_run-validation-tag-epoch_accuracy.csv')
plt.figure(figsize=(15,10))
plt.xlabel('Epochs')
plt.ylabel('accuracy')
plt.plot(train_acc_1vs3['Value'], label='training')
plt.plot(val_acc_1vs3['Value'], label='validation')
plt.legend(loc='lower right')
plt.show() 
```

```python

train_loss_1vs3 = pd.read_csv('train_time_data/1vs3_run-train-tag-epoch_loss.csv')
val_loss_1vs3 = pd.read_csv('train_time_data/1vs3_run-validation-tag-epoch_loss.csv')
plt.figure(figsize=(15,10))
plt.xlabel('Epochs')
plt.ylabel('loss')
plt.plot(train_loss_1vs3['Value'], label='training')
plt.plot(val_loss_1vs3['Value'], label='validation')
plt.legend(loc='lower right')
plt.show()
```
